```python
## Creating income bands using custom bins
bins=[0, 100000, 300000 , float('inf')]
labels=['Low', 'Medium', 'High']

df["Income Band"]= pd.cut(df["Estimated Income"], bins=bins, labels=labels, include_lowest=True)
```

```python
numerical_cols = ['Fee Structure','Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans',
                   'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
## Visualizing distribution of numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()

```

```python
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']


correlation_matrix = df[numerical_cols].corr()

## Visualizing the correlatin matrix using heatmap
plt.figure(figsize=(12, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]


for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(8, 6))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},   ## Making points semi_transparent  
        line_kws={'color': 'red'}     ## best fit line color  
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

Insights

(1). Deposits and Savings Behavior

- Strong correlation between bank deposits and savings accounts suggests overlapping behaviors (e.g., total funds kept in the bank).

- Customers who deposit regularly also tend to grow their savings balances.

(2). Income, Age, and Accumulation

- Age and estimated income show moderate correlations with superannuation, savings, and checking balances.

- Higher earners and older individuals often accumulate larger savings, retirement funds, and may carry higher credit card or loan balances.

(3). Low Correlation with Property Ownership

- Property ownership depends on external factors (e.g., housing markets, inheritance, location).

- These drivers are not captured by banking variables, leading to weaker correlations.

(4). Business vs. Personal Banking

- Business lending has a moderate relationship with bank loans, indicating some overlap between personal and business borrowing.

- Its weak connection to deposits or property metrics suggests it serves a distinct customer segment or purpose.
